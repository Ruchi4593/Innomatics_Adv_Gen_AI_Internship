##Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
highest_score = max(employees.values())

# Find all employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

##Problem Statement 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
import string
for p in string.punctuation:
    query = query.replace(p, "")

# Split sentence into words
words = query.split()

# Count frequency
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display only keywords searched more than once
result = {}

for key, value in freq.items():
    if value > 1:
        result[key] = value

print(result)
```

##Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create an empty list to store valid (even) readings
valid_readings = []

# enumerate() is used to get both index (hour) and value (reading)
for index, value in enumerate(sensor_readings):

    # Check if the reading is even (valid)
    if value % 2 == 0:

        # Store the valid reading as a tuple (hour_index, reading_value)
        valid_readings.append((index, value))

# Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

##Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store count of each email domain
domain_count = {}

# Total number of emails
total_emails = len(emails)

# Loop through each email in the list
for email in emails:

    # Extract the domain part after '@'
    domain = email.split("@")[1]

    # Count the occurrence of each domain
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate and display percentage usage of each domain
for domain, count in domain_count.items():

    # Calculate percentage
    percentage = (count / total_emails) * 100

    # Display result
    print(domain + ":", str(int(percentage)) + "%")
```

##Problem Statement 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Calculate adjusted threshold (approx 30% above average)
threshold = average_sales * 1.29

# Detect spike days
for index, value in enumerate(sales):

    # Check if sales exceed or equal the threshold
    if value >= threshold:
        print("Day", index + 1, ":", value)
```

##Problem Statement 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store count of each user ID
id_count = {}

# Count occurrence of each user ID
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Display only duplicate user IDs (appearing more than once)
for uid, count in id_count.items():

    # Check if the user ID appears more than once
    if count > 1:
        print(uid, "→", count, "times")
```
