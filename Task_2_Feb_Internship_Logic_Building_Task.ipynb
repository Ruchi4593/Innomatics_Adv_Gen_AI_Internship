##Problem 1: Unique Words in a Sentence
###Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
sentence = input("Enter sentence: ")

# Split the sentence into a list of words
words = sentence.split()

# Convert the list of words to a set to automatically remove duplicates
unique_words = set(words)

# Get the count of unique words
unique_count = len(unique_words)

# Print the count of unique words
print(f"Unique words count: {unique_count}")

# Print the set of unique words
print(f"Unique words: {unique_words}")
```

##Problem 2: Highest Salary from Employee Data
###Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.


```python
# Define the dictionary with employee names as keys and salaries as values
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000 }

# Initialize variables to track the highest salary and the corresponding employee
highest_employee = None
highest_salary = 0

# Loop through the dictionary to find the employee with the highest salary
for name, salary in employees.items():
    if salary > highest_salary:  # Checking if current salary is higher
        highest_salary = salary  # Update the highest salary
        highest_employee = name  # Update the employee name

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

##Problem 3: Find Maximum and Minimum Values
###Problem Statement
Find the maximum and minimum values from a list of numbers.


```python
numbers = [45, 22, 89, 10, 66]

# Print the original list
print(f"List: {numbers}")

# Initialize max and min with the first element of the list
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list starting from the second element to find max and min
for num in numbers[1:]:
    if num > max_value:  # Update max if current number is larger
        max_value = num
    if num < min_value:  # Update min if current number is smaller
        min_value = num

# Print the maximum and minimum values
print(f"Max: {max_value}")
print(f"Min: {min_value}")
```

##Problem 4: Count Products Above a Price Threshold
###Problem Statement
Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]

# Initialize a counter to keep track of products above 1000
count = 0

# Loop through each price in the list
for price in prices:
    if price > 1000:  # Check if the price is greater than 1000
        count += 1  # Increment the counter if condition is met

# Print the count of products above 1000
print(f"Products above 1000: {count}")
```

##Problem 5: Calculate Attendance Percentage
###Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P"]

# total number of days
total_days = len(attendance)

# Initialize a counter for present days
present_count = 0

# Loop through the attendance list to count "P" (Present)
for status in attendance:
    if status == "P":  # Check if the status is Present
        present_count += 1  # Increment the counter

# Calculate the attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Print the attendance percentage
print(f"Attendance Percentage: {attendance_percentage:.1f}")
```

##Problem 7: Count Character Frequency
###Problem Statement
Count how many times each character appears in a given string.


```python
input_string = "python"

# Initialize an empty dictionary to store character frequencies
char_frequency = {}

# Loop through each character in the string
for char in input_string:
    # Check if the character is already in the dictionary
    if char in char_frequency:
        # If yes, increment its count
        char_frequency[char] += 1
    else:
        # If not, add it to the dictionary with a count of 1
        char_frequency[char] = 1

print(char_frequency)
```

##Problem 10: Calculate Average Marks
###Problem Statement
Calculate the average marks of a student from a list of marks.


```python
marks = [70, 75]

# Calculate the sum of all marks in the list
total_sum = sum(marks)

# Get the total number of marks
total_marks = len(marks)

# Calculate the average by dividing the sum by the total number of marks
average_marks = total_sum / total_marks

# Print the average marks, formatted to one decimal place
print(f"Average Marks: {average_marks:.1f}")
```
