#Problem 1: Mobile Recharge Validation System
##Problem Statement
Create a function to validate a mobile recharge request.



```python
# Function to validate recharge request
def validate_recharge(amount, valid_plans):

    # Check if amount is less than ₹50
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Check if amount is in valid plans
    elif amount not in valid_plans:
        print("Invalid plan selected. Available plans are:", valid_plans)
        return False

    # If both conditions are satisfied
    else:
        print("Recharge Successful!")
        return True


# List of valid recharge plans
valid_plans = [199, 299, 399, 599]

# Retry option using while loop
while True:

    # Take recharge amount as input
    amount = int(input("Enter recharge amount: ₹"))

    # Call validation function
    if validate_recharge(amount, valid_plans):
        break
    else:
        print("Please try again.\n")
```

#Problem 2: Inventory Reorder Alert System
##Problem Statement
###Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


```python
# Function to check inventory stock levels
def check_inventory(stock_data):

    # Loop through each product and its quantity in the dictionary
    for product, quantity in stock_data.items():

        # Check if stock is less than 15
        if quantity < 15:
            print(product, "→ Reorder Alert")

        # Otherwise stock is sufficient
        else:
            print(product, "→ Stock OK")


# Dictionary containing product names and their stock quantities
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18,
    "Printer": 8
}

# Call the function
check_inventory(inventory)
```

#Problem 3: Student Result Processing System
##Problem Statement
###Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


```python
# Function to process student result
def process_result(marks):

    # Variable to store total marks
    total = 0

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    # Calculate average marks
    average = total / len(marks)

    # Check pass or fail condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# List of student marks
student_marks = [60, 45, 70, 55, 40]

# Call function and store result
result = process_result(student_marks)

# Display result
print("Result:", result)
```

#Problem 4: Cab Fare Estimator with Retry Option
##Problem Statement
###Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):

    # Base fare
    fare = 50 + (12 * distance)

    # Add 25% extra during peak hours
    if peak_hour == "yes":
        fare = fare + (0.25 * fare)

    return fare


# Retry option using while loop
while True:

    # Take distance input
    distance = float(input("Enter distance travelled (in km): "))

    # Ask if it is peak hour
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    # Calculate fare using function
    total_fare = calculate_fare(distance, peak_hour)

    # Display fare
    print("Total Cab Fare: ₹", int(total_fare))

    # Ask user if they want to retry
    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        break
```

#Problem 5: Employee Attendance Eligibility Checker
##Problem Statement
###Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


```python
# Function to check attendance eligibility
def check_eligibility(attendance):

    # Variable to count present days
    present_count = 0

    # Loop to count number of "P"
    for day in attendance:
        if day == "P":
            present_count += 1

    # Calculate attendance percentage
    percentage = (present_count / len(attendance)) * 100

    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Attendance list of employee
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function and store result
result = check_eligibility(attendance_list)

# Display result
print("Attendance Status:", result)
```

#Problem 6: Password Strength Checker
##Problem Statement
###Create a function to check password strength.
####Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


```python
# Function to check password strength
def check_password(password):

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    # Loop through each character in password
    for ch in password:

        # Check for digit
        if ch.isdigit():
            has_digit = True

        # Check for special character
        if ch in "@#$":
            has_special = True

    # Check all conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit & special character)"


# Take password input from user
user_password = input("Enter Password: ")

# Call function and display result
result = check_password(user_password)
print(result)
```
