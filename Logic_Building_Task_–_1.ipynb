##1. User Login Check
###Given a username and password, check whether login is successful.


```python
#predefined username and password
username = "admin"
password = "1234"

input_user = input("Enter username: ")
input_password = input("Enter password: ")

#Checking if username and password matches with the predefined username and password
if input_user == username and input_password == password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

##2. Pass / Fail Analyzer
###Analyze student results.

```python
marks = [45, 78, 90, 33, 60]

# Initialize counters for pass and fail students
pass_count = 0  #Counter for students who passed
fail_count = 0  #Counter for students who failed

#using loop to compare pass and fail students
for mark in marks:
    if mark >= 50:
        pass_count += 1  # Increment pass counter
    else:
        fail_count += 1  # Increment fail counter

total_students = len(marks)

print("Pass / Fail Analysis Results:")
print(f"Total Students: {total_students}")
print(f"Number of Students Passed: {pass_count}")
print(f"Number of Students Failed: {fail_count}")

```

##3. Simple Data Cleaner
###Clean and standardize user names.


```python
names = [" Alice ", "bob", " CHARLIE "]

# Created an empty list to store clean names
cleaned_names = []

#using loop to process each name in the list
for name in names:
  # For each name, remove any extra spaces at the beginning or end using strip()
  # Then, convert it to all lowercase using lower() to standardize it
  cleaned_name = name.strip().lower()

  cleaned_names.append(cleaned_name)

print(cleaned_names)
```

##4. Message Length Analyzer
###Analyze message sizes.


```python
messages = ["Hi", "Welcome to the platform", "OK"]

#using loop to check length in each message
for message in messages:
  length = len(message) #calculate the length of message

  print(f"Message: '{message}' - Length: {length} characters")

  #checking length if greater than 10 characters
  if length > 10:
    print(" This message is longer than 10 characters.")

  print()
```

##5. Error Message Detector
###Detect error messages from system logs.


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

#initializing a counter to count how many "ERROR" entries are in the list
error_count = 0

#using loop to check for each log entry in the list to check for errors
for log in logs:
  if log == "ERROR": #checking if log entry is "ERROR"
    error_count += 1 #If yes increament the error count

#print total number of "ERROR" count
print(f"Total number of ERROR entries: {error_count}")
```
