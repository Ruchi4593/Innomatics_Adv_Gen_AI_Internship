##Problem 1: Smart Light Controller
###Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
# Define the SmartLight class
class SmartLight:
    # Constructor: set the light's name and default state to OFF
    def __init__(self, name):
        self.name = name          # store the light's name
        self.is_on = False       # initially the light is OFF

    # Turn the light on
    def turn_on(self):
        self.is_on = True        # set the state to ON

    # Turn the light off
    def turn_off(self):
        self.is_on = False       # set the state to OFF

    # Return the status string
    def status(self):
        state = "ON" if self.is_on else "OFF"   # choose the word
        return f"{self.name} light is {state}"         # format the message


# Main code with prompts
if __name__ == "__main__":
    # Ask for light name with a visible prompt
    light_name = input("Enter Light Name: ")

    # Ask for action with a visible prompt
    action = input("Enter Action (ON/OFF): ")

    # Create a SmartLight object
    light = SmartLight(light_name)

    # Apply the requested action
    if action == "ON":
        light.turn_on()          # turn the light on
    elif action == "OFF":
        light.turn_off()         # turn the light off

    # Print the current status
    print(light.status())
```

##Problem 2: Employee ID Card System
###Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.


```python
# Define the Employee class
class Employee:
    # Constructor: initialize employee name, ID, and department
    def __init__(self, name, emp_id, department):
        self.name = name                    # store employee name
        self.emp_id = emp_id                # store employee ID
        self.department = department        # store department

    # Return the ID card details as a formatted string
    def display_id_card(self):
        # Create the ID card format
        card = f"----- Employee ID Card -----\n" \
               f"Name: {self.name}\n" \
               f"ID: {self.emp_id}\n" \
               f"Department: {self.department}\n" \
               f"--------------------"
        return card


# Main code with prompts
if __name__ == "__main__":
    # Get employee details from user with visible prompts
    name = input("Enter Employee Name: ")
    emp_id = input("Enter Employee ID: ")
    department = input("Enter Department: ")

    # Create an Employee object
    employee = Employee(name, emp_id, department)

    # Display the ID card
    print(employee.display_id_card())
```

##Problem 3: Mobile Contact Record
###Problem Statement
Create a class to store contact name and phone number and display contact information.


```python
# Define the Contact class
class Contact:
    # Constructor: initialize contact name and phone number
    def __init__(self, name, phone):
        self.name = name          # store contact name
        self.phone = phone        # store phone number

    # Display the contact information
    def display(self):
        print("Contact Saved")    # confirmation message
        print(f"Name: {self.name}")     # print name
        print(f"Phone: {self.phone}")    # print phone number


# Main code with prompts
if __name__ == "__main__":
    # Get contact details from user with visible prompts
    name = input("Enter Contact Name: ")
    phone = input("Enter Phone Number: ")

    # Create a Contact object
    contact = Contact(name, phone)

    # Display the contact information
    contact.display()
```

##Problem 4: Product Price Tag Generator
###Problem Statement
Create a class that stores product name and price and prints a formatted price tag.


```python
# Define the Product class
class Product:
    # Constructor: initialize product name and price
    def __init__(self, name, price):
        self.name = name          # store product name
        self.price = price        # store product price

    # Print the formatted price tag
    def display_price_tag(self):
        # Print product name with label
        print(f"Product: {self.name}")
        # Print price with rupee symbol
        print(f"Price: ₹{self.price}")


# Main code with prompts
if __name__ == "__main__":
    # Get product details from user
    name = input("Enter Product Name: ")
    price = input("Enter Price: ")

    # Create a Product object
    product = Product(name, price)

    # Display the price tag
    product.display_price_tag()
```

##Problem 5: Movie Rating Display System
###Problem Statement
Create a class that stores movie name and rating and displays movie details.


```python
# Define the Movie class
class Movie:
    # Constructor: initialize movie name and rating
    def __init__(self, name, rating):
        self.name = name          # store movie name
        self.rating = rating      # store movie rating

    # Display the movie details
    def display(self):
        # Print movie name with label
        print(f"Movie: {self.name}")
        # Print rating with "/ 5" format
        print(f"Rating: {self.rating} / 5")


# Main code with prompts
if __name__ == "__main__":
    # Get movie details from user
    name = input("Enter Movie Name: ")
    rating = input("Enter Rating: ")

    # Create a Movie object
    movie = Movie(name, rating)

    # Display the movie details
    movie.display()
```

##Problem 6: Delivery Address Manager
###Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.


```python
# Define the Delivery class
class Delivery:
    # Constructor: initialize customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name    # store customer name
        self.address = address                # store delivery address

    # Display the delivery details
    def display(self):
        # Print delivery details header
        print("Delivery Details")
        # Print customer name with label
        print(f"Customer: {self.customer_name}")
        # Print address with label
        print(f"Address: {self.address}")


# Main code with prompts
if __name__ == "__main__":
    # Get delivery details from user
    customer_name = input("Enter Customer Name: ")
    address = input("Enter Address: ")

    # Create a Delivery object
    delivery = Delivery(customer_name, address)

    # Display the delivery details
    delivery.display()
```
