##Problem Statement 1: Smart Parking Lot Management System
Design a function to manage a smart parking lot.


```python
def smart_parking_manager(capacity: int, logs: list) -> dict:
    current = 0          # vehicles currently inside
    peak   = 0           # highest count ever seen
    alert  = False       # becomes True if we ever exceed capacity

    for movement in logs:
        move = movement.strip().upper()   # normalise input

        if move == "IN":
            current += 1
            # update peak if we just broke the record
            if current > peak:
                peak = current
            # check for over‑capacity
            if current > capacity:
                alert = True
        elif move == "OUT":
            # A car leaves – we never go below zero (defensive)
            current = max(0, current - 1)
        else:
            pass

    # Decide the textual status
    if alert:
        status = "Alert: Exceeds Capacity!"
    elif current == capacity:
        status = "Full"
    else:
        status = "Available"

    return {
        "currently_parked": current,
        "peak_usage":       peak,
        "status":           status
    }

if __name__ == "__main__":
    PARKING_CAPACITY = 50
    VEHICLE_LOGS = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

    result = smart_parking_manager(PARKING_CAPACITY, VEHICLE_LOGS)

    print(f"Currently Parked Vehicles: {result['currently_parked']}")
    print(f"Parking Status: {result['status']}")
    print(f"Peak usage (max cars at any moment): {result['peak_usage']}")
```

##Problem Statement 2:Online Food Delivery Time Estimator
###Create a function that estimates delivery time based on:
Distance (km)


Weather condition


Traffic level


```python
def calculate_delivery_time(distance: float, traffic_level: str, weather: str) -> dict:
    # Base speed: 4 minutes per kilometer
    base_time = distance * 4

    # Traffic multipliers
    traffic_mult = {"Low": 1.0, "Medium": 1.15, "High": 1.25}
    mult = traffic_mult.get(traffic_level.capitalize(), 1.0)
    time_traffic = base_time * mult

    # Weather delays
    weather_delay = {"Clear": 0, "Rainy": 10, "Foggy": 15, "Snowy": 20}
    delay = weather_delay.get(weather.capitalize(), 0)

    # Final time with prep time
    final_time = round(time_traffic + delay + 5)

    return {
        "base_time": base_time,
        "traffic_mult": mult,
        "weather_delay": delay,
        "final_time": final_time
    }


# Display formatted output
def display_eta(result: dict, distance: float, traffic: str, weather: str):
    print("=" * 50)
    print("        FOOD DELIVERY TIME ESTIMATOR")
    print("=" * 50)
    print(f"Distance: {distance} km")
    print(f"Traffic: {traffic}")
    print(f"Weather: {weather}")
    print("-" * 50)
    print(f"Base Time: {result['base_time']} min")
    print(f"Traffic ({result['traffic_mult']}x): +{round(result['base_time'] * (result['traffic_mult'] - 1))} min")
    print(f"Weather Delay: +{result['weather_delay']} min")
    print(f"Prep Time: +5 min")
    print("-" * 50)
    print(f"ESTIMATED DELIVERY TIME: {result['final_time']} minutes")
    print("=" * 50)


# Sample input from problem statement
distance = 8
traffic_level = "High"
weather = "Rainy"

# Calculate and display
result = calculate_delivery_time(distance, traffic_level, weather)
display_eta(result, distance, traffic_level, weather)

```

##Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
###Build a function that analyzes seat booking data and:
Calculates occupancy percentage


Determines if show is Housefull


Suggests opening additional shows


```python
def analyze_occupancy(total_seats: int, booked_seats: list) -> dict:
    # Count booked seats from the list
    booked_count = sum(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status based on occupancy
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Half Full"
    elif occupancy >= 25:
        status = "Few Seats"
    else:
        status = "Empty"

    # Suggest additional shows if occupancy is high
    if occupancy >= 80:
        suggestion = "Consider opening additional shows"
    else:
        suggestion = "Current show capacity is sufficient"

    return {
        "booked_count": booked_count,
        "total_seats": total_seats,
        "occupancy": round(occupancy, 2),
        "status": status,
        "suggestion": suggestion
    }


# Display results
def display_occupancy(result: dict):
    print("=" * 50)
    print("     MOVIE THEATRE OCCUPANCY ANALYZER")
    print("=" * 50)
    print(f"Total Seats: {result['total_seats']}")
    print(f"Booked Seats: {result['booked_count']}")
    print("-" * 50)
    print(f"Occupancy: {result['occupancy']}%")
    print(f"Show Status: {result['status']}")
    print(f"Suggestion: {result['suggestion']}")
    print("=" * 50)


# Sample input from problem statement
TOTAL_SEATS = 200
# Create list with 150 booked seats (1) and 50 empty seats (0)
BOOKED_SEATS = [1] * 150 + [0] * 50

# Analyze occupancy
result = analyze_occupancy(TOTAL_SEATS, BOOKED_SEATS)

# Display results
display_occupancy(result)
```

##Problem Statement 4:Cloud Server Load Classification System
###Create a function to classify server load based on CPU usage readings.
####Rules:
Average CPU < 50% → Normal


50%–80% → Warning


80% → Critical


```python
def classify_server_load(cpu_readings: list) -> dict:
    # Calculate average CPU usage
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    # Classify server status based on average
    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    return {
        "avg_cpu": round(avg_cpu, 2),
        "status": status,
        "readings_count": len(cpu_readings)
    }


# Display results
def display_load(result: dict, readings: list):
    print("=" * 50)
    print("   CLOUD SERVER LOAD CLASSIFICATION SYSTEM")
    print("=" * 50)
    print(f"CPU Readings: {readings}")
    print(f"Number of Readings: {result['readings_count']}")
    print("-" * 50)
    print(f"Average CPU Load: {result['avg_cpu']}%")
    print(f"Server Status: {result['status']}")
    print("=" * 50)


# Sample input from problem statement
CPU_READINGS = [45, 60, 70, 85, 90]

# Classify server load
result = classify_server_load(CPU_READINGS)

# Display results
display_load(result, CPU_READINGS)
```

##Problem Statement 5:Smart Classroom Resource Usage Monitor
###Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

```python
# Define thresholds for each resource (in hours)
THRESHOLDS = {
    "Projector": 8,
    "AC": 8,
    "Lights": 8
}


def analyze_resource_usage(usage: dict) -> dict:
    # Find overused resources
    overused = []
    for resource, hours in usage.items():
        threshold = THRESHOLDS.get(resource, 8)
        if hours > threshold:
            overused.append(resource)

    # Check if energy alert is needed
    alert = "Yes" if len(overused) > 0 else "No"

    return {
        "overused": overused,
        "alert": alert,
        "usage": usage
    }


# Display results
def display_usage(result: dict):
    print("=" * 50)
    print("   SMART CLASSROOM RESOURCE USAGE MONITOR")
    print("=" * 50)
    print(f"Resource Usage: {result['usage']}")
    print(f"Thresholds: {THRESHOLDS}")
    print("-" * 50)

    overused_str = ", ".join(result['overused']) if result['overused'] else "None"
    print(f"Overused Resources: {overused_str}")
    print(f"Energy Alert: {result['alert']}")
    print("=" * 50)


# Sample input from problem statement
RESOURCE_USAGE = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Analyze resource usage
result = analyze_resource_usage(RESOURCE_USAGE)

# Display results
display_usage(result)
```

##Problem Statement 6: Online Event Registration Capacity Controller
###Create a function that manages event registrations by:
Tracking registrations


Preventing overbooking


Triggering waitlist mode


```python
def manage_registrations(capacity: int, total_registrations: int) -> dict:
    # Calculate confirmed and waitlisted
    confirmed = min(capacity, total_registrations)
    waitlisted = max(0, total_registrations - capacity)

    # Determine registration status
    if total_registrations >= capacity:
        status = "Closed"
    elif total_registrations >= capacity * 0.9:
        status = "Almost Full"
    else:
        status = "Open"

    return {
        "confirmed": confirmed,
        "waitlisted": waitlisted,
        "status": status,
        "capacity": capacity,
        "total_registrations": total_registrations
    }


# Display results
def display_registration(result: dict):
    print("=" * 50)
    print("   EVENT REGISTRATION CAPACITY CONTROLLER")
    print("=" * 50)
    print(f"Event Capacity: {result['capacity']}")
    print(f"Total Registrations: {result['total_registrations']}")
    print("-" * 50)
    print(f"Confirmed Registrations: {result['confirmed']}")
    print(f"Waitlisted Users: {result['waitlisted']}")
    print(f"Registration Status: {result['status']}")
    print("=" * 50)


# Sample input from problem statement
EVENT_CAPACITY = 100
TOTAL_REGISTRATIONS = 105

# Manage registrations
result = manage_registrations(EVENT_CAPACITY, TOTAL_REGISTRATIONS)

# Display results
display_registration(result)
```
